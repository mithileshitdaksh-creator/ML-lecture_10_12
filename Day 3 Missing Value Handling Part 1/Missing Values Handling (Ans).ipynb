# Handling the missing values.

##### Drop all the rows having missing value.

##### - Python and Pandas consider nan as missing values.

```python
df.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
df
```

```python
df.dropna()
```

##### Delete row if entire that row having missing values.

```python
df.dropna(how="all")
```

##### Delete the row if Rating, year, collection all of them have missing value in some row.

```python
df.dropna(how="Any", subset=["Rating",	"Collection", "Year of Rel."])
```

```python
df
```

##### Delete the row if Rating, year, collection all of them have missing value in same row.

```python
df.dropna(how="all", subset=["Rating",	"Collection", "Year of Rel."])
```

# Fill the missing values.

##### Replace the NAN of rating to 0

```python
df["Rating"].fillna(value=0)
```

```python
df["Rating"] = df["Rating"].fillna(value=0)
```

```python
df
```

##### Raplace or fill the na value of Year with it's mean.

```python
df["Year of Rel."].mean()
```

```python
df["Year of Rel."].fillna(df["Year of Rel."].mean())
```
