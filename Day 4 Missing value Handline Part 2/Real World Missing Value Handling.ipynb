### 13. Now actucally handle the missing value.

- Step 1 : # Replace the missing value with NAN

```python
df["normalized-losses"] = df["normalized-losses"].replace(to_replace="?", value=np.nan)
```

- Step 2 : # Change the datatype of that column.

```python
df["normalized-losses"] = df["normalized-losses"].astype("float")
```

- Step 3 : # find the mean of that column.

```python
means = df['normalized-losses'].mean()
```

- Step 4 : # Fill all the NAN with mean.

```python
df['normalized-losses'] = df['normalized-losses'].fillna(means)
```
